```python
data1=pd.read_csv("Hyderabad.csv")
```

```python
data2=pd.read_csv("Bangalore.csv")
```

```python
data4=pd.read_csv(r"C:\Users\yuges\Chennai.csv")
```

```python
data6=pd.read_csv(r"C:\Users\yuges\Pune.csv")
```

```python
data7=pd.read_csv(r"C:\Users\yuges\Kolkata.csv")
```

```python
data8=pd.read_csv(r"C:\Users\yuges\New Delhi.csv")
```

```python
data9=pd.read_csv(r"C:\Users\yuges\Lucknow.csv")
```

```python
data10=pd.read_csv(r"C:\Users\yuges\Noida.csv")
```

```python
data=pd.concat([data1,data2,data3,data4,data5,data6,data7,data8,data9,data10],axis=0)
```

```python
data
```

```python
data.isna().sum() #Checking for missing values
```

```python
data["Review (in words)"].replace(np.nan,"No Review",inplace=True) #Replacing the missing values with No Review
```

```python
data["Review (in words)"].isna().sum()
```

```python
data["Number of Reviews"].replace(np.nan,0,inplace=True) #Filling the missing values in Number of Reviews column with 0
```

```python
data["Number of Reviews"].isna().sum()
```

```python
data["Rating (out of 10)"].replace(np.nan,0.0,inplace=True) #Filling the missing values in Rating column with 0.0
```

```python
data["Rating (out of 10)"].isna().sum()
```

```python
data.isna().sum()
```

```python
data["Level of sustainability"].replace(np.nan,0,inplace=True) #Replacing missing values in Level of sustainability column with 0
```

```python
data["Level of sustainability"].isna().sum()
```

```python
data["Type of Cancellation"].replace(np.nan,"Not Free",inplace=True) #Missing values are filled with Not Free in Type of Cancellation column
```

```python
data["Type of Cancellation"].isna().sum()
```

```python
data["Taxes and charges"].replace(np.nan,0,inplace=True) #Replacing missing values with 0
```

```python
data["Taxes and charges"].isna().sum()
```

```python
data["Type of Payment"].replace(np.nan,"prepayment",inplace=True) #Replacing missing values with prepayment
```

```python
data["Type of Payment"].isna().sum()
```

```python
data["Type of Bed"].fillna(data["Type of Bed"].mode()[0],inplace=True) #Filling missing values with mode because of categorical data
```

```python
data.isna().sum() #All missing values are filled 
```

```python
data
```

```python
data["Cost"]=data["Cost"].replace("₹|,","",regex=True) #Removing special characters in Cost column
```

```python
data["Taxes and charges"]=data["Taxes and charges"].replace("₹|,","",regex=True) #Removing special characters
```

```python
data
```

```python
data["Number of Reviews"]=data["Number of Reviews"].replace(",|review|reviews|external|s","",regex=True) #Removing review,reviews and ,
```

```python
data
```

```python
data["Level of sustainability"]=data["Level of sustainability"].replace("Level| ","",regex=True)
```

```python
data
```

```python
data["Distance from Centre (in km)"]=data["Distance from Centre (in km)"].replace("km","",regex=True)
```

```python
data
```
