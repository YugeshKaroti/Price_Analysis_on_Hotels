```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Hotels.csv")
```

```python
df
```

```python
sns.histplot(df["Distance from Centre (in km)"],kde=True,color="Blue")
plt.title("Histogram for Distance from Centre (in km) column")
plt.show()
```

```python
plt.style.available
```

```python
plt.style.use("ggplot")
sns.histplot(df["Rating (out of 10)"],kde=True,color="Red")
plt.title("Histogram for Rating (out of 10) column")
plt.show()
```

```python
sns.histplot(df["Cost"],kde=True,color="green")
plt.title("Histogram for Cost column")
plt.show()
```

```python
sns.histplot(df["Taxes and charges"],kde=True,color="Indigo")
plt.title("Histogram for Taxes and charges column")
plt.show()
```

```python
df["City"].value_counts().plot.pie(autopct="%.2f")
plt.title("Pie chart for different cities")
plt.show()
```

```python
df["Review (in words)"].value_counts().plot.bar(color=["red","green","orange","purple","blue","maroon","Indigo"])
plt.xlabel("Reviews")
plt.ylabel("Count")
plt.title("Bar plot for Reviews (in words) column")
plt.show()
```

```python
df.columns
```

```python
sns.countplot(x="Level of sustainability",data=df,width=0.3)
plt.title("Count plot for Level of sustainability column")
plt.show()
```

```python
plt.style.use("ggplot")
sns.countplot(x="Type of Cancellation",data=df,width=0.3)
plt.title("Count plot for Type of Cancellation column")
plt.show()
```

```python
sns.countplot(x="Type of Payment",data=df,width=0.2)
plt.title("Count plot for Type of Payment column")
plt.show()
```

```python
plt.scatter(df["Rating (out of 10)"],df["Cost"],marker="*",color="blue")
plt.xlabel("Rating")
plt.ylabel("Price")
plt.title("Scatter plot for Rating and Cost")
plt.show()
```
