```python
Hotel_name=[]
Location=[]
Distance_from_city=[]
Overall_review=[]
Number_of_reviews=[]
Overall_rating=[]
Sustainable_level=[]
Room_Type=[]
Bed_Type=[]
Free_Cancellation=[]
price=[]
Tax=[]
prepayment=[]
for i in range(1,100,25):
    url=f"https://www.booking.com/searchresults.en-gb.html?label=msn-SvOYT_f6pES7XcBmICjudw-80401890429994%3Atikwd-17116145739%3Aaud-807859093%3Aloc-90%3Aneo%3Amte%3Alp158516%3Adec%3Aqshotels%2Bin%2Bmumbai&sid=af0d52358b8d524164a8d6d5aca460cb&aid=339462&ss=Hyderabad&ssne=Hyderabad&ssne_untouched=Hyderabad&lang=en-gb&sb=1&src_elem=sb&src=searchresults&dest_id=-2097701&dest_type=city&checkin=2023-09-28&checkout=2023-09-29&group_adults=1&no_rooms=1&group_children=0&offset={i}"
    driver.get(url)
    soup=BeautifulSoup(driver.page_source,"html.parser")
    for i in soup.find_all("div",class_="b978843432"):
        a=i.find("div",class_="f6431b446c a23c043802")
        if(a):
            Hotel_name.append(a.text)
        else:
            Hotel_name.append(np.nan)
    for i in soup.find_all("div",class_="b978843432"):
        text=i.text
        regex=re.findall("\d+.?\d?\skm",text)
        if(regex):
            Distance_from_city.append(regex[0])
        else:
            Distance_from_city.append(np.nan)
    for i in soup.find_all("div",class_="b978843432"):
        a=i.find("div",class_="a3b8729ab1 e6208ee469 cb2cbb3ccb")
        if(a):
            Overall_review.append(a.text)
        else:
            Overall_review.append(np.nan)
    for i in soup.find_all("div",class_="b978843432"):
        b=i.find("div",class_="abf093bdfe f45d8e4c32 d935416c47")
        if(b):
            Number_of_reviews.append(b.text)
        else:
            Number_of_reviews.append(np.nan)
    for i in soup.find_all("div",class_="b978843432"):
        a=i.find("div",class_="a3b8729ab1 d86cee9b25")
        if(a):
            Overall_rating.append(a.text)
        else:
            Overall_rating.append(np.nan)
    for i in soup.find_all("div",class_="b978843432"):
        text=i.text
        regex=re.findall("\s(Level\s\d)",text)
        if(regex):
            Sustainable_level.append(regex[0])
        else:
            Sustainable_level.append(np.nan)
    for i in soup.find_all("div",class_="b978843432"):
        a=i.find("span",class_="df597226dd")
        if(a):
            Room_Type.append(a.text)
        else:
            Room_Type.append(np.nan)
    for i in soup.find_all("div",class_="b978843432"):
        a=i.find("div",class_="cb5b4b68a4")
        if(a):
            Bed_Type.append(a.text)
        else:
            Bed_Type.append(np.nan)
    for i in soup.find_all("div",class_="d7449d770c a081c98c6f"):
        a=i.find("span",class_="f6431b446c fbd1d3018c e729ed5ab6")
        if(a):
            price.append(a.text)
        else:
            price.append(np.nan)
    for i in soup.find_all("div",class_="b978843432"):
        e=i.find("div",class_="abf093bdfe d068504c75")
        if(e):
            Free_Cancellation.append(e.text)
        else:
            Free_Cancellation.append(np.nan)
    for i in soup.find_all("div",class_="b978843432"):
        text=i.text
        regex=re.findall("\+(â‚¹\s\d+,?\d+)",text)
        if(regex):
            Tax.append(regex[0])
        else:
            Tax.append(np.nan)
    for i in soup.find_all("div",class_="b978843432"):
        text=i.text
        regex=re.findall("No\sprepayment\sneeded",text)
        if(regex):
            prepayment.append(regex[0])
        else:
            prepayment.append(np.nan)
city="Hyderabad"
for i in range(1,101):
    Location.append(city)
```

#### DataFrame for hotels in Hyderabad

```python
dict1={"Hotel Name":Hotel_name,"City":Location,"Distance from City":Distance_from_city,"Review (in words)":Overall_review,"Number of Reviews":Number_of_reviews,"Rating (out of 10)":Overall_rating,"Level of sustainability":Sustainable_level,"Type of Room":Room_Type,"Number of Beds":Bed_Type,"Type of Cancellation":Free_Cancellation,"Cost":price,"Taxes and charges":Tax,"Type of Payment":prepayment}
```
